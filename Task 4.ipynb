## 1

```r
library("openxlsx")
library("dplyr")
```

```r
df <- read.xlsx("./data_input/GAZ.xlsx")
head(df)
```

Полям загруженных данных даём чуть более понятные названия

```r
colnames(df) <- c("Дата замера", "Давление", "Температура", "Добыча", "Конденсат", "Вода", "ID", "Куст", "Группа")
head(df)
```

Дальше идёт преобразование разных форматов даты в человеческий. Если с форматом *d/m/Y* всё легко, то с датами по типу *42343* приходится использовать дополнительные библиотеки. Но всё равно каким-то образом в глубинах таблички создаются **NA**...

```r
df[1] <- sapply(df[1], function(x) { ifelse(is.na(as.numeric(x)), as.character(as.Date(x, format="%d/%m/%Y")), as.character(convertToDate(x, origin = "1900-01-01"))) })
head(df)
```

```r
typeof(df$ID)
```

Далее и до конца пункта идёт преобразование типов полей в нужные

```r
df$ID <- factor(df$ID)
df$Куст <- factor(df$Куст)
df$Группа <- factor(df$Группа)
```

```r
head(df)
tail(df)
```

```r
df <- transform(df, Давление = as.numeric(Давление))
df <- transform(df, Температура = as.numeric(Температура))
df <- transform(df, Добыча = as.numeric(Добыча))
df <- transform(df, Конденсат = as.numeric(Конденсат))
df <- transform(df, Вода = as.numeric(Вода))
head(df)
```

## 2

Очистка от строк с пропущенными значениями

```r
df <- df[rowSums(is.na(df)) == 0,]
head(df)
```

## 3

Преобразование температуры из шкалы Цельсия в Кельвины

```r
df <- transform(df, Температура = Температура + 273.15)
head(df)
```

## 5

В этом пункте создаются безразмерные поля по заданию:

* Отношение добычи газа к конденсату;
* Отношение добычи газа к добыче воды;
* Отношение добычи воды к добыче конденсата.

При их создании в некоторых случаях образуются значения **NA** и **Inf**, что будет учтено далее

!! Alert: Inf

```r
df["НефтиНаКонденсат"] <- df["Добыча"] / df["Конденсат"]
head(df)
```

!! Alert: Inf

```r
df["НефтиНаВоду"] <- df["Добыча"] / df["Вода"]
head(df)
```

!! Alert: Inf or NA

```r
df["ВодыНаКонденсат"] <- df["Вода"] / df["Конденсат"]
head(df)
```

## 6

```r
getYear <- function(dateString) {
    return (as.numeric(format(as.Date(dateString),'%Y')))
}
```

```r
df_2018 <- filter(df, getYear(Дата.замера) == 2018)
head(df_2018)
```

```r
mining <- aggregate(Добыча ~ Куст, df_2018, sum)
mining
```

```r
right_results <- filter(df, df$НефтиНаВоду != Inf)
kusts <- aggregate(НефтиНаВоду ~ Куст, right_results, mean)
kusts
```

Название куста, в котором среднее отношение добычи нефти к добыче воды было самым наибольшим за всё время наблюдений

```r
kusts[order(kusts$НефтиНаВоду, decreasing=TRUE), ][1, "Куст"]
```
